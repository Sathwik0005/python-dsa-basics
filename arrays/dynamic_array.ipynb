```python
import ctypes

class MyList:

  def __init__(self):
    self.size = 1
    self.n = 0
    self.A = self.__make_array(self.size)

  def __len__(self):
    return self.n

  def __resize(self, new_capacity):
    B = self.__make_array(new_capacity)
    self.size = new_capacity

    for i in range(self.n):
      B[i] = self.A[i]

    self.A = B

  def append(self,item):
    if self.size == self.n:
      self.__resize(self.size*2)
    self.A[self.n] = item
    self.n = self.n + 1

  def pop(self):
    if self.n == 0:
      return 'List is Empty'
    print(self.A[self.n-1])
    self.n = self.n - 1

  def clear(self):
    self.n = 0
    self.size = 1

  def find(self,item):
    for i in range(self.n):
      if self.A[i] == item:
        return i
    return 'Value Error - not in list'

  def __str__(self):
    result = ''
    for i in range(self.n):
      result = result + str(self.A[i]) + ','
    return '[' + result[:-1] + ']'

  def __getitem__(self, index):
    if 0<= index < self.n:
      return self.A[index]
    else:
      return 'IndexError - Index out of range'

  def __delitem__(self, pos):
    if 0 <= pos < self.n:
      for i in range(pos, self.n-1):
        self.A[i] = self.A[i+1]
      self.n = self.n - 1

  def remove(self, item):
    pos = self.find(item)
    if type(pos) == int:
      self.__delitem__(pos)
    else:
      return pos

  def __make_array(self, capacity):
    return (capacity*ctypes.py_object)()

  def insert(self,pos,item):
    if self.n == self.size:
      self.__resize(self.size*2)

    for i in range(self.n, pos, -1):
      self.A[i] = self.A[i-1]
    self.A[pos] = item
    self.n += 1

  def sort(self):
    for i in range(self.n):
      for j in range(i+1, self.n):
        if self.A[i] > self.A[j]:
          temp = self.A[i]
          self.A[i] = self.A[j]
          self.A[j] = temp

  def min(self):
    min = self.A[0]
    for i in range(1,self.n):
      if min > self.A[i]:
        min = self.A[i]
    return min

  def max(self):
    max = self.A[0]
    for i in range(1,self.n):
      if max < self.A[i]:
        max = self.A[i]
    return max

  def sum(self):
    sum = 0
    for i in range(self.n):
      sum = sum + self.A[i]
    return sum

  def extend(self, list):
    list_len = len(list)
    if self.size < list_len:
      self.__resize(self.size*2)
    else:
      for i in list:
        self.append(i)

  def neg_index(self, pos):
    index = self.n + pos
    if 0 <= index < self.n:
      return self.A[index]
    else:
      return 'IndexError - Index out of range'

  def slice(self, start=None, stop=None, step=1):
    if start is None:
      start = 0
    if stop is None:
      stop = self.n

    if start < 0:
      start = self.n + start

    if stop < 0:
      stop = self.n + stop

    my_array = []

    for i in range(start, stop, step):
      my_array.append(self.A[i])

    return my_array

  def merge(self,other):
    new_array = []

    for i in range(self.n):
      new_array.append(self.A[i])

    for i in range(len(other)):
      new_array.append(other[i])

    return new_array
```

```python
print("---- Creating list ----")
arr = MyList()                     # 1. create list
print("Initial length:", len(arr)) # 2. len
```

```python
print("\n---- Append & Print ----")
arr.append(10)
arr.append(5)
arr.append(20)                     # 3. append
print("Array:", arr)               # 4. print (__str__)
```

```python
print("\n---- Indexing ----")
print("arr[0] =", arr[0])          # 5. indexing
print("arr[1] =", arr[1])
```

```python
print("\n---- Pop ----")
arr.pop()                          # 6. pop (prints last element)
print("After pop:", arr)
```

```python
print("\n---- Insert ----")
arr.insert(1, 99)                  # 9. insert at position 1
print("After insert at index 1:", arr)
```

```python
print("\n---- Delete by index ----")
del arr[0]                         # 10. delete (__delitem__)
print("After deleting index 1:", arr)
```

```python
print("\n---- Remove by value ----")
arr.append(30)
arr.append(40)
print("Before remove(30):", arr)
arr.remove(30)                     # 11. remove
print("After remove(30):", arr)
```

```python
print("\n---- Find ----")
print("Index of 10:", arr.find(10))     # 8. find
print("Index of 100:", arr.find(100))   # not found
```

```python
print("\n---- Min / Max / Sum / Sort ----")
print("Min:", arr.min())
print("Max:", arr.max())
print("Sum:", arr.sum())
arr.sort()                         # sort
print("After sort:", arr)
```

```python
print("\n---- Extend ----")
arr.extend([7, 8, 9])              # extend
print("After extend [7, 8, 9]:", arr)
```

```python
print("\n---- Negative Indexing ----")
print("Last element (neg_index(-1)):", arr.neg_index(-1))
print("Second last (neg_index(-2)):", arr.neg_index(-2))
```

```python
print("\n---- Slicing ----")
print("slice(0, 3):", arr.slice(0, 3))
print("slice(1, None):", arr.slice(1, None))
print("slice(-3, None):", arr.slice(-3, None))
print("slice(0, len, 2):", arr.slice(0, len(arr), 2))
```

```python
print("\n---- Merge ----")
other = [100, 200, 300]
merged = arr.merge(other)
print("Merged with [100, 200, 300]:", merged)
```

```python
print("\n---- Clear ----")
arr.clear()                        # 7. clear
print("After clear, length:", len(arr))
print("Array:", arr)
```
